```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/content/drive/MyDrive/MLproject/datasets/preprocessed_data.csv')
```

```python
masked_data = pd.read_csv('/content/drive/MyDrive/MLproject/datasets/masked_dataset.csv')
```

```python
data = data.dropna()
```

```python
data.info()
```

here we are executing for simple dataset without subemtion...

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data.text,data.is_depression , test_size = 0.30)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer(ngram_range=(1,2))
X_train_cv = v.fit_transform(X_train.values)
X_test_cv = v.transform(X_test.values)
type(X_train_cv)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report
# Create an SVM classifier
svm_classifier = SVC(kernel='linear')

# Train the SVM classifier
svm_classifier.fit(X_train_cv, y_train)

# Make predictions on the test set
y_pred = svm_classifier.predict(X_test_cv)
```

```python
# Calculate accuracy
print(classification_report(y_test,y_pred))

from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
