```python
import pandas as pd
data = pd.read_csv('/content/drive/MyDrive/MLproject/datasets/preprocessed_data.csv')
masked_data = pd.read_csv('/content/drive/MyDrive/MLproject/datasets/masked_dataset.csv')
data.shape
```

```python
data = data.dropna()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data.text,data.is_depression , test_size = 0.2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
X_train_cv = v.fit_transform(X_train.values)
type(X_train_cv)
```

```python
X_train_cv.toarray()[0]
```

```python
v.get_feature_names_out()[100:200]
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_cv,y_train)
```

```python
X_test_cv = v.transform(X_test)
```

```python
from sklearn.metrics import classification_report
y_pred = model.predict(X_test_cv)
print(classification_report(y_test,y_pred))
```

Now use masked data for depression detection...........

```python

X_train, X_test, y_train, y_test = train_test_split(masked_data.masked_text,masked_data.is_depression , test_size = 0.2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v = CountVectorizer()
X_train_cv = v.fit_transform(X_train.values)
type(X_train_cv)
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_cv,y_train)
```

```python
X_test_cv = v.transform(X_test)
```

```python
from sklearn.metrics import classification_report
y_pred = model.predict(X_test_cv)
print(classification_report(y_test,y_pred))
```
